# Urban Heat Island Classification Challenge
## A Machine Learning Approach Using Sentinel-2 Satellite Imagery and Building Footprint Data



## 1. Introduction

Urban Heat Islands (UHI) are a growing environmental challenge in which densely developed urban areas experience significantly higher temperatures than their surrounding rural regions — sometimes by as much as 10 degrees Celsius. This phenomenon arises from the replacement of natural land cover with heat-absorbing impervious surfaces such as concrete, asphalt, and rooftops, combined with the waste heat generated by transportation, industry, and air conditioning. The consequences are serious: extreme heat is the leading weather-related cause of death globally, and UHI effects disproportionately harm the most vulnerable populations, including young children, the elderly, outdoor workers, and low-income communities who lack access to cooling infrastructure.

This challenge tasked us with developing a machine learning classification model to predict UHI intensity — categorized as Low, Medium, or High — using geospatial satellite and building footprint data. Training data consisted of ground-measured Urban Heat Index values collected via vehicle and bicycle traversals across two cities: Santiago, Chile (January 2024) and Rio de Janeiro, Brazil (January 2023). Our model was then applied to predict UHI intensity across 14,105 locations in Freetown, Sierra Leone — an unseen city in a different geographic and climatic context entirely.

The primary data sources were Sentinel-2 multispectral satellite imagery from the European Space Agency's Copernicus mission and three-dimensional building footprint data from the 3D-GloBFP dataset. Sentinel-2 provides global imagery across 12 spectral bands at 10–60 metre resolution, enabling the computation of spectral indices that characterize vegetation, water, bare soil, and urban surfaces. The 3D-GloBFP dataset adds structural urban information — building count, footprint area, and coverage — that captures the physical geometry of urban fabric. Together these datasets allowed our model to identify the spectral and structural signatures associated with different levels of urban heat intensity, and to generalize those patterns from South American cities to a West African city without any labeled data from that region.


## 2. Recommendations

### 2a. Which Factor Contributes Most to Predicting Urban Heat Islands?

Based on our SHAP (feature importance) analysis, the single most powerful predictor of urban heat island intensity was the **Redness Index of the soil at the regional 200-metre scale** — specifically, how much bare, exposed soil exists in the area surrounding a measurement point. In plain terms: where the ground is exposed, unshaded, and uncovered by vegetation or buildings, it heats up dramatically under direct sunlight and radiates that heat back into the surrounding environment. This finding makes intuitive sense in the context of rapidly growing cities like Freetown and Rio de Janeiro, where urban expansion frequently strips away natural vegetation without immediately replacing it with buildings, leaving large stretches of exposed ground that act as heat radiators.

Close behind in importance were measures of **water presence and variability at the 200-metre scale**, particularly the Modified Normalized Difference Water Index. Proximity to water bodies — rivers, coastlines, ponds — provides a strong cooling effect through evaporation, and the spatial variability of water signals in the surrounding area tells the model how fragmented or continuous that cooling influence is. Locations surrounded by consistent water signals tend to have low UHI intensity, while those where water is scarce or patchy experience significantly more heat stress.

A practical takeaway from these findings is that city planners should prioritize **covering exposed soil surfaces** as a first-order intervention. This does not require expensive infrastructure — even low-cost ground cover such as gravel, mulch, shade cloth, or drought-tolerant ground-cover plants can dramatically reduce the heat-absorbing capacity of bare surfaces. In areas where immediate planting is not feasible, light-coloured reflective coatings on exposed ground can reduce heat absorption at very low cost. The data consistently showed that the 200-metre regional context matters more than the immediate point-level surface — meaning that isolated improvements to a single lot have less impact than neighbourhood-scale strategies that address the surrounding 200 metres of land cover comprehensively.

---

### 2b. Practical Solutions to Reduce Urban Heat Islands (Beyond Green Space)

While planting trees and creating parks are well-known UHI mitigation strategies, our analysis points to several additional practical interventions that can meaningfully reduce urban heat intensity in Santiago, Rio de Janeiro, and Freetown.

**Cool roofs and reflective pavements** represent one of the most cost-effective interventions available. Our model found that SWIR (shortwave infrared) band variability — a proxy for how much heat different surfaces absorb and emit — was among the top predictors of UHI intensity. Dark rooftops and asphalt absorb up to 95% of incoming solar radiation. Coating rooftops white or with reflective materials, and using light-coloured concrete or permeable pavers for roads and pavements, can reduce surface temperatures by 10–20°C and lower air temperatures in surrounding areas. This is particularly actionable in dense neighbourhoods where vegetation cannot easily be planted.

**Water infrastructure and blue corridors** are a second high-impact intervention. The strong predictive power of water indices in our model confirms what urban climatologists have long argued: waterways, retention ponds, and water features actively cool surrounding areas through evaporation. Cities can amplify this effect by restoring degraded waterways, creating stormwater retention features that hold water rather than draining it immediately, and designing urban water features along streets and public spaces. In Freetown specifically, which is coastal, improving access to and preservation of natural waterways running from the hills to the coast could provide significant and geographically widespread cooling.

**Building design and urban geometry** also emerged as important factors. Our building morphology features — particularly coverage ratio and building count — showed that the density and arrangement of buildings influences heat retention. Narrow streets flanked by tall buildings trap heat in canyon-like geometries. Cities can address this through zoning regulations that require minimum setbacks, open courtyards, and building orientations that allow prevailing winds to ventilate urban corridors. In new development areas, requiring passive cooling design standards — including natural ventilation, thermal mass, and shading — can prevent heat island formation before it occurs rather than requiring costly remediation afterward.

**Reducing waste heat from transportation** is a fourth practical lever. Our index-based features captured the spectral signature of impervious road surfaces, which both absorb heat and generate it through vehicle traffic. Transitioning public transportation to electric vehicles, creating pedestrian and cycling corridors that reduce car traffic in dense urban cores, and implementing congestion pricing in high-density areas all reduce the waste heat deposited directly into urban environments. These interventions have co-benefits including reduced air pollution and improved public health beyond the UHI effect alone.

### 2c. Additional Data That Would Add Value to UHI Prediction

Given unlimited time, three publicly available datasets would significantly improve the predictive power and generalizability of our UHI model.

**Land Surface Temperature (LST) from Landsat or MODIS** would be the single most valuable addition. NASA's Landsat 8/9 and Terra/Aqua MODIS satellites provide thermal infrared measurements of actual surface temperature at 30–1000 metre resolution, globally, dating back to the 1970s. Currently our model predicts UHI class from reflectance-based spectral proxies — it infers temperature from land cover type. Direct surface temperature measurements would allow the model to learn the relationship between spectral indices and actual thermal conditions, producing more physically grounded and accurate predictions. MODIS LST data is freely available through NASA's Earth Data portal.

**Digital Elevation Models (DEM) and terrain derivatives** would improve predictions significantly in topographically complex cities. Freetown is built on a peninsula with dramatic elevation changes — coastal flatlands, steep hillsides, and valley corridors. Cold air pools in valleys, heat accumulates on south-facing slopes, and elevation strongly modulates temperature through the adiabatic lapse rate. The SRTM (Shuttle Radar Topography Mission) provides 30-metre global elevation data for free, and derivatives such as slope, aspect, topographic wetness index, and solar radiation exposure could all be computed from it. Our current model has no terrain information and therefore cannot distinguish a coastal flat from a hillside — both may have similar spectral signatures but very different thermal behavior.

**Nighttime light intensity from VIIRS or DMSP-OLS** would add a proxy for human activity and waste heat generation that is not captured by daytime satellite imagery. The Visible Infrared Imaging Radiometer Suite (VIIRS) Day/Night Band provides global nighttime radiance data at 750-metre resolution, freely available through NOAA. Brighter nighttime lights indicate denser commercial and industrial activity, which generates substantial waste heat from lighting, HVAC systems, and machinery. This data is particularly valuable for distinguishing between areas with similar daytime land cover but very different levels of human activity — a satellite industrial park and a dense residential neighbourhood may look similar in Sentinel-2 reflectance but produce vastly different amounts of waste heat at night.

## 3. Analysis and Code

Our modeling pipeline was designed to maximize predictive performance while maintaining physical interpretability and cross-regional generalizability. The core challenge was building a model that learns UHI patterns from South American cities and applies them reliably to a West African city with different climate, vegetation, and urban morphology — a classic domain transfer problem in geospatial machine learning.

**Feature Engineering** was the most critical component of our pipeline. Rather than sampling a single pixel at each measurement point (as in the baseline model), we extracted features from three concentric bounding-box windows — 50 metres, 100 metres, and 200 metres — around each location. For each window we computed four statistics: mean, standard deviation, 25th percentile, and 75th percentile. This multi-scale approach captures both the immediate surface type at a point and its broader neighbourhood context, which our results confirmed is more predictive than point-level sampling alone. The 200-metre scale features consistently dominated in SHAP importance, confirming that regional land-use context drives UHI intensity more than local surface type.

We computed 21 spectral indices from the 12 raw Sentinel-2 bands, covering vegetation health (NDVI, EVI, SAVI, MSAVI, GNDVI, NDVIre), urban built-up intensity (NDBI, UI, IBI, BRBA, BCI, NDISI), soil exposure (BSI, RI), water presence (NDWI, MNDWI, AWEInsh), and heat stress (NBR, NBR2). A key technical challenge was that the EVI formula contains a denominator that can approach zero for certain tropical pixel combinations, producing astronomically large values that destroyed cross-region scaling. We resolved this by clamping the EVI denominator to a minimum of 0.001 and clipping all index outputs to physically valid ranges before extraction. We also excluded atmosphere-only bands (B01 Coastal Aerosol and B09 Water Vapour), which our analysis showed were picking up systematic differences in atmospheric conditions between training and validation regions rather than genuine land surface signals.

**Cross-domain interaction features** were engineered by multiplying spectral indices with building morphology metrics — for example, NDVI multiplied by building density captures the presence of green space inside dense urban areas, while NDBI multiplied by volume density amplifies the urban heat signal in three-dimensionally dense blocks. Urban-vegetation contrast features such as `NDBI − NDVI` directly quantify the dominance of built-up surfaces over vegetation at each location. These 25 engineered features collectively capture physical interactions that neither the satellite nor the building data can express alone.

**Model Selection** used an ensemble of three gradient boosting models: two LightGBM instances with different configurations (one Optuna-tuned, one with fixed deeper architecture) and one sklearn GradientBoostingClassifier. Soft-voting — averaging predicted class probabilities across all three models — combined their strengths and reduced individual model errors. LightGBM's hyperparameters were tuned using Optuna Bayesian optimization over 50 trials with 5-fold stratified cross-validation, finding optimal settings including a learning rate of 0.196, 698 estimators, and a minimum leaf sample size of 45. The high minimum leaf sample size is particularly important for cross-region generalization, as it prevents the model from learning overly specific patterns from rare training examples.

**SHAP-based feature selection** was applied to rank all features by their contribution to predictions and remove the bottom 20% by importance. In practice, the full feature set (396 features) slightly outperformed the pruned set (316 features), so the full set was retained. **Pseudo-labeling** was attempted — high-confidence Sierra Leone predictions were added back to the training set to help the model adapt to the target region — but a distribution safety guard rejected the pseudo-labeled models when they produced fewer than 5% Medium predictions, correctly identifying that the pseudo-labels were skewing the output. The final ensemble achieved a weighted F1 score of **0.9200** on the held-out test set, with per-class F1 scores of 0.94 for High, 0.94 for Low, and 0.88 for Medium.


<a id='conclusion'></a>
## 4. Conclusion

This challenge demonstrated that satellite remote sensing data, combined with building footprint information and thoughtful feature engineering, can predict urban heat island intensity with high accuracy — achieving an F1 score of 0.92, more than double the 0.40 baseline. The most important lesson from our modeling work is that regional context matters more than local surface type: the 200-metre neighbourhood scale consistently outperformed the 50-metre local scale in feature importance, meaning that the thermal environment of a location is shaped more by what surrounds it than by what is directly beneath it.

Our SHAP analysis revealed that bare soil exposure and water index variability — not building density, as one might initially expect — are the strongest predictors of UHI intensity. This finding has direct practical implications: covering exposed soil and preserving water access are among the most cost-effective interventions available to city planners, particularly in rapidly developing cities in the Global South where resources are constrained. Building cool roofs, maintaining blue corridors, and designing for natural ventilation are complementary strategies that do not require the years-long growth cycles associated with tree planting.

The cross-regional transfer from South American training cities to a West African validation city presented significant technical challenges, most notably the inconsistency of atmospheric correction across different climate zones. Future work should incorporate Land Surface Temperature measurements from Landsat or MODIS thermal bands, terrain data from SRTM, and nighttime light intensity from VIIRS — all freely available datasets that would meaningfully improve both predictive accuracy and physical interpretability. As 70% of the world's population is projected to live in urban areas by 2050, tools like this model will become increasingly essential for identifying where heat islands are forming, understanding why they form, and targeting interventions where they will have the greatest impact on human health and wellbeing.

```python
# ── STEP 3: Band Map & Spectral Index Engine ──────────────────────────────────
# 12-band GeoTIFF: B01,B02,B03,B04,B05,B06,B07,B08,B8A,B09,B11,B12
# Key fixes vs baseline:
#   - EVI denominator clamped (prevents billion-scale values in tropical areas)
#   - All indices clipped to physically valid ranges
#   - AWEInsh normalised by mean band value (not a ratio — needs normalisation)
#   - Atmosphere bands excluded (B01/CAI, B09/WVI hurt cross-region generalisation)

BAND_IDX = {
    'B01':1,'B02':2,'B03':3,'B04':4,'B05':5,'B06':6,
    'B07':7,'B08':8,'B8A':9,'B09':10,'B11':11,'B12':12,
}
EPS = 1e-10

def safe_clip(arr, lo, hi):
    return np.clip(arr, lo, hi)

def compute_all_indices(bands):
    B02=bands['B02']; B03=bands['B03']; B04=bands['B04']
    B05=bands.get('B05'); B06=bands.get('B06')
    B08=bands['B08']; B8A=bands.get('B8A')
    B11=bands['B11']; B12=bands.get('B12')
    idx = {}

    # Vegetation
    idx['NDVI']  = safe_clip((B08-B04)/(B08+B04+EPS), -1, 1)
    evi_d = B08+6*B04-7.5*B02+1
    evi_d = np.where(np.abs(evi_d)<0.001, 0.001, evi_d)
    idx['EVI']   = safe_clip(2.5*(B08-B04)/evi_d, -1, 1)
    idx['SAVI']  = safe_clip(1.5*(B08-B04)/(B08+B04+0.5+EPS), -1.5, 1.5)
    idx['MSAVI'] = safe_clip((2*B08+1-np.sqrt(np.maximum((2*B08+1)**2-8*(B08-B04),0)))/2,-1,1)
    idx['GNDVI'] = safe_clip((B08-B03)/(B08+B03+EPS),-1,1)
    if B05 is not None:
        idx['NDVIre'] = safe_clip((B08-B05)/(B08+B05+EPS),-1,1)
        idx['CIre']   = safe_clip((B08/(B05+EPS))-1,-10,10)
    if B06 is not None and B05 is not None:
        idx['RENDVI'] = safe_clip((B06-B05)/(B06+B05+EPS),-1,1)

    # Urban
    idx['NDBI'] = safe_clip((B11-B08)/(B11+B08+EPS),-1,1)
    idx['UI']   = safe_clip((B11-B8A)/(B11+B8A+EPS) if B8A is not None else (B11-B08)/(B11+B08+EPS),-1,1)
    ibi_n = 2*B11/(B11+B08+EPS)-(B08/(B08+B04+EPS)+B03/(B03+B11+EPS))
    ibi_d = 2*B11/(B11+B08+EPS)+(B08/(B08+B04+EPS)+B03/(B03+B11+EPS))+EPS
    idx['IBI']  = safe_clip(ibi_n/ibi_d,-1,1)
    idx['BRBA'] = safe_clip(B04/(B11+EPS),0,10)
    idx['BCI']  = safe_clip(((B11+B04)/2-B08)/((B11+B04)/2+B08+EPS),-1,1)
    if B12 is not None:
        idx['NDISI'] = safe_clip((B12-(B08+B03)/2)/(B12+(B08+B03)/2+EPS),-1,1)

    # Soil
    idx['BSI'] = safe_clip(((B11+B04)-(B08+B02))/((B11+B04)+(B08+B02)+EPS),-1,1)
    idx['RI']  = safe_clip((B04-B03)/(B04+B03+EPS),-1,1)

    # Water
    idx['NDWI']  = safe_clip((B03-B08)/(B03+B08+EPS),-1,1)
    idx['MNDWI'] = safe_clip((B03-B11)/(B03+B11+EPS),-1,1)
    if B12 is not None:
        awei = 4*(B03-B11)-(0.25*B08+2.75*B12)
        bscale = (B03.mean()+B11.mean()+B08.mean()+B12.mean())/4+EPS
        idx['AWEInsh'] = safe_clip(awei/bscale,-20,20)

    # Burn/heat stress
    idx['NBR']  = safe_clip((B08-B11)/(B08+B11+EPS),-1,1)
    if B12 is not None:
        idx['NBR2'] = safe_clip((B11-B12)/(B11+B12+EPS),-1,1)

    # Raw bands (excluding atmosphere-only B01, B09)
    EXCL = {'B01','B09'}
    for bname, barr in bands.items():
        if bname not in EXCL:
            idx[f'raw_{bname}'] = safe_clip(barr/10000.0, 0, 3)

    return {k:v for k,v in idx.items() if v is not None}

print('Band map and spectral index engine ready')
```

```python
# ── STEP 4: Multi-Scale Spatial Feature Extractor ─────────────────────────────
# Samples bounding-box windows at 50m, 100m, and 200m around each point.
# Computes mean, std, p25, p75 for every spectral index per window.
# Output naming: <index>_<stat>_<buffer>m  e.g. NDVI_mean_100m

BUFFER_SIZES = [50, 100, 200]
STATS = ['mean', 'std', 'p25', 'p75']

def _stat(arr, stat):
    a = arr[np.isfinite(arr)]
    if len(a) == 0: return np.nan
    return {'mean':np.mean,'std':np.std,
            'p25':lambda x:np.percentile(x,25),
            'p75':lambda x:np.percentile(x,75)}[stat](a)

def extract_multiscale_sentinel_features(geotiff_path, csv_path,
                                          buffer_sizes=BUFFER_SIZES, stats=STATS):
    df = pd.read_csv(csv_path)
    lats, lons = df['Latitude'].values, df['Longitude'].values
    all_rows = []
    with rasterio.open(geotiff_path) as src:
        n_bands = src.count
        print(f'   Band count: {n_bands}')
        for lat, lon in tqdm(zip(lats,lons), total=len(lats),
                             desc=f'  {os.path.basename(geotiff_path)}'):
            row_feats = {}
            for buf_m in buffer_sizes:
                half_deg = buf_m/111320.0
                try:
                    window = from_bounds(lon-half_deg,lat-half_deg,
                                        lon+half_deg,lat+half_deg, src.transform)
                    raw = src.read(window=window).astype(np.float32)
                    raw[raw<=0] = np.nan
                except:
                    raw = np.full((n_bands,1,1), np.nan, dtype=np.float32)
                band_arrays = {bname: raw[bidx-1].ravel()
                               for bname,bidx in BAND_IDX.items() if bidx<=n_bands}
                indices = compute_all_indices(band_arrays)
                for idx_name, vals in indices.items():
                    if not hasattr(vals,'__len__'): vals=np.array([vals])
                    for stat in stats:
                        row_feats[f'{idx_name}_{stat}_{buf_m}m'] = _stat(vals, stat)
            all_rows.append(row_feats)
    return pd.concat([df.reset_index(drop=True), pd.DataFrame(all_rows)], axis=1)

print('Multi-scale extractor ready')
```

```python
# ── STEP 5: Building Morphology Feature Extractor ─────────────────────────────
# Computes 9 building features within a 100m square buffer around each point.

def compute_building_features(csv_path, shp_path, lat_col='Latitude',
                               lon_col='Longitude', buffer_m=100):
    df = pd.read_csv(csv_path)
    df['geometry'] = df.apply(lambda r: Point(r[lon_col],r[lat_col]), axis=1)
    gdf_pts  = gpd.GeoDataFrame(df, geometry='geometry', crs='EPSG:4326')
    gdf_bldg = gpd.read_file(shp_path)
    if str(gdf_bldg.crs) != 'EPSG:3857': gdf_bldg = gdf_bldg.to_crs(epsg=3857)
    gdf_pts  = gdf_pts.to_crs(epsg=3857)
    height_col = next((c for c in gdf_bldg.columns
                       if any(k in c.lower() for k in ('height','hgt','h_','ht_','elev'))), None)
    print(f'   Height column: {height_col or "not found — volume = 0"}')
    buf_area = float(buffer_m*buffer_m)
    records  = []
    for _, row in tqdm(gdf_pts.iterrows(), total=len(gdf_pts), desc='  Building features'):
        pt  = row.geometry
        buf = box(pt.x-buffer_m/2,pt.y-buffer_m/2,pt.x+buffer_m/2,pt.y+buffer_m/2)
        bldgs = gdf_bldg[gdf_bldg.geometry.intersects(buf)].copy()
        bldgs['geometry'] = bldgs.geometry.buffer(0)
        n_bldgs = len(bldgs)
        if n_bldgs > 0:
            clipped = bldgs.copy()
            clipped['geometry'] = clipped.geometry.intersection(buf)
            clipped = clipped[~clipped.geometry.is_empty]
            total_area = clipped.geometry.area.sum()
            coverage   = total_area/buf_area
            if height_col:
                heights = pd.to_numeric(bldgs[height_col],errors='coerce').dropna()
                mean_h = heights.mean() if len(heights)>0 else 0.0
                max_h  = heights.max()  if len(heights)>0 else 0.0
                std_h  = heights.std()  if len(heights)>1 else 0.0
                vol_h  = pd.to_numeric(bldgs.loc[clipped.index,height_col],
                                       errors='coerce').fillna(mean_h)
                volume = (clipped.geometry.area*vol_h.values).sum()
            else:
                mean_h=max_h=std_h=volume=0.0
        else:
            total_area=coverage=mean_h=max_h=std_h=volume=0.0
        records.append({'building_count':n_bldgs,
                        'building_density_100m':n_bldgs/buf_area,
                        'total_footprint_area_m2':total_area,
                        'coverage_ratio':coverage,
                        'mean_height_m':mean_h,'max_height_m':max_h,
                        'std_height_m':std_h,'total_volume_m3':volume,
                        'volume_density':volume/buf_area})
    result = gdf_pts.reset_index(drop=True)
    for col in records[0]: result[col]=[r[col] for r in records]
    return result

BLDG_COLS = ['building_count','building_density_100m','total_footprint_area_m2',
             'coverage_ratio','mean_height_m','max_height_m','std_height_m',
             'total_volume_m3','volume_density']
print('Building extractor ready')
```

```python
# ── STEP 6: Extraction for All Three Regions ──────────────────────────────
# NOTE: This step takes approximately 3-4hours mins to complete.
# Sentinel extraction: ~35 minutes total
# Building extraction: ~3 hours total (Chile is slowest due to dense shapefile)

print('\n  Sentinel-2 feature extraction...')
print('Chile:')
chile_sat  = extract_multiscale_sentinel_features(GEOTIFF_CHILE,  CSV_CHILE)
print('Brazil:')
brazil_sat = extract_multiscale_sentinel_features(GEOTIFF_BRAZIL, CSV_BRAZIL)
print('Sierra Leone:')
sierra_sat = extract_multiscale_sentinel_features(GEOTIFF_SIERRA, CSV_SIERRA)

SAT_FEAT_COLS = [c for c in chile_sat.columns
                 if c not in ('Latitude','Longitude','UHI_Class','geometry')]
print(f'\n Sentinel extraction done — {len(SAT_FEAT_COLS)} features')

print('\n Building feature extraction...')
print('Chile:')
chile_bldg  = compute_building_features(CSV_CHILE,  SHP_CHILE)
print('Brazil:')
brazil_bldg = compute_building_features(CSV_BRAZIL, SHP_BRAZIL)
print('Sierra Leone:')
sierra_bldg = compute_building_features(CSV_SIERRA, SHP_SIERRA)
print(f'\n Building extraction done — {len(BLDG_COLS)} features')
```

```python
# ── STEP 7: Merge & Feature Engineering ──────────────────────────────────────

def merge_region(sat_df, bldg_df, sat_cols, bldg_cols, label_col='UHI_Class'):
    out = sat_df[sat_cols].copy().reset_index(drop=True)
    for c in bldg_cols:
        if c in bldg_df.columns: out[c] = bldg_df[c].values
    if label_col in sat_df.columns: out[label_col] = sat_df[label_col].values
    return out

def engineer_features(df):
    out = df.copy()
    EPS = 1e-10
    def g(name):
        col = f'{name}_mean_100m'
        return out[col] if col in out.columns else pd.Series(np.nan, index=out.index)
    ndvi=g('NDVI'); ndbi=g('NDBI'); ndwi=g('NDWI'); mndwi=g('MNDWI')
    bsi=g('BSI'); ibi=g('IBI'); nbr=g('NBR'); ui=g('UI')
    bdens  = out.get('building_density_100m',  pd.Series(0.0,index=out.index))
    cov    = out.get('coverage_ratio',         pd.Series(0.0,index=out.index))
    volden = out.get('volume_density',         pd.Series(0.0,index=out.index))
    mean_h = out.get('mean_height_m',          pd.Series(0.0,index=out.index))
    bcount = out.get('building_count',         pd.Series(0.0,index=out.index))
    tot_a  = out.get('total_footprint_area_m2',pd.Series(0.0,index=out.index))
    # Cross-domain interactions
    out['feat_ndvi_x_bdens']  = ndvi*bdens
    out['feat_ndbi_x_volden'] = ndbi*volden
    out['feat_ndwi_x_cov']    = ndwi*cov
    out['feat_bsi_x_cov']     = bsi*cov
    out['feat_ui_x_meanh']    = ui*mean_h
    out['feat_ibi_x_volden']  = ibi*volden
    out['feat_ndvi_x_meanh']  = ndvi*mean_h
    out['feat_mndwi_x_bdens'] = mndwi*bdens
    # Urban-vegetation contrast
    out['feat_urban_veg_diff']  = ndbi-ndvi
    out['feat_urban_veg_ratio'] = (ndbi+EPS)/(ndvi.abs()+EPS)
    out['feat_heat_stress_idx'] = ndbi-ndwi
    out['feat_cool_cover']      = ndvi+mndwi
    out['feat_nbr_x_bdens']     = nbr*bdens
    # Multi-scale gradients
    for n in ['NDVI','NDBI','NDWI','BSI','MNDWI']:
        m50=out.get(f'{n}_mean_50m',pd.Series(np.nan,index=out.index))
        m100=out.get(f'{n}_mean_100m',pd.Series(np.nan,index=out.index))
        m200=out.get(f'{n}_mean_200m',pd.Series(np.nan,index=out.index))
        s100=out.get(f'{n}_std_100m',pd.Series(np.nan,index=out.index))
        out[f'feat_{n}_scale_grad'] = m200-m50
        out[f'feat_{n}_cv']         = s100/(m100.abs()+EPS)
    # Building shape
    out['feat_bldg_avg_size'] = tot_a/(bcount+EPS)
    out['feat_bldg_aspect']   = mean_h/(np.sqrt(tot_a/(bcount+EPS))+EPS)
    return out

chile_fe  = engineer_features(merge_region(chile_sat,  chile_bldg,  SAT_FEAT_COLS, BLDG_COLS))
brazil_fe = engineer_features(merge_region(brazil_sat, brazil_bldg, SAT_FEAT_COLS, BLDG_COLS))
sierra_fe = engineer_features(merge_region(sierra_sat, sierra_bldg, SAT_FEAT_COLS, BLDG_COLS))

train_fe = pd.concat([chile_fe, brazil_fe], ignore_index=True)
val_fe   = sierra_fe.copy()
ENG_COLS = [c for c in train_fe.columns if c.startswith('feat_')]
print(f'Training set: {train_fe.shape} | Validation: {val_fe.shape}')
print(f'Engineered features: {len(ENG_COLS)}')
```

```python
# ── STEP 8: Preprocessing ─────────────────────────────────────────────────────

FEATURE_COLS = [c for c in train_fe.columns
                if c not in ('UHI_Class','Latitude','Longitude','geometry')]

# Drop constants
FEATURE_COLS = [c for c in FEATURE_COLS
                if train_fe[c].nunique()>1 and train_fe[c].notna().sum()>10]

# Cross-region divergence check: remove columns where val deviates > 20 train stds
train_means = train_fe[FEATURE_COLS].mean()
train_stds  = train_fe[FEATURE_COLS].std()
val_means   = val_fe[FEATURE_COLS].mean()
z_scores    = (val_means-train_means).abs()/(train_stds+1e-10)
bad_cols    = z_scores[z_scores>20].index.tolist()
print(f'Columns removed by divergence check: {len(bad_cols)}')
FEATURE_COLS = [c for c in FEATURE_COLS if c not in bad_cols]
print(f'Features after divergence check: {len(FEATURE_COLS)}')

# Deduplicate
train_fe = train_fe.drop_duplicates(subset=FEATURE_COLS, keep='first').reset_index(drop=True)

# Clip outliers (fit on train only)
clip_lo = train_fe[FEATURE_COLS].quantile(0.005)
clip_hi = train_fe[FEATURE_COLS].quantile(0.995)
train_fe[FEATURE_COLS] = train_fe[FEATURE_COLS].clip(lower=clip_lo, upper=clip_hi, axis=1)
for c in FEATURE_COLS:
    if c in val_fe.columns: val_fe[c] = val_fe[c].clip(clip_lo[c], clip_hi[c])

# NaN imputation with training median
medians = train_fe[FEATURE_COLS].median()
train_fe[FEATURE_COLS] = train_fe[FEATURE_COLS].fillna(medians)
for c in FEATURE_COLS:
    if c in val_fe.columns: val_fe[c] = val_fe[c].fillna(medians[c])

# Encode labels
le = LabelEncoder()
y  = le.fit_transform(train_fe['UHI_Class'])
print(f'Classes: {le.classes_}  →  {np.unique(y)}')

X = train_fe[FEATURE_COLS].values
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.20, stratify=y, random_state=42)

# Single consistent scaler
scaler  = StandardScaler()
X_tr    = scaler.fit_transform(X_tr)
X_te    = scaler.transform(X_te)
X_val   = scaler.transform(val_fe[FEATURE_COLS].values)

print(f'X_tr mean: {X_tr.mean():.4f} | X_val mean: {X_val.mean():.4f}')
print(f'X_train: {X_tr.shape} | X_test: {X_te.shape} | X_val: {X_val.shape}')
print('Preprocessing complete')
```

```python
# ── STEP 9: Hyperparameter Tuning with Optuna ─────────────────────────────────
# NOTE: This step takes approximately 3-4 hours to complete.
N_TRIALS = 50
CV_FOLDS = 5
skf = StratifiedKFold(n_splits=CV_FOLDS, shuffle=True, random_state=42)

def cv_f1(model, X, y):
    scores = []
    for tr_i, vl_i in skf.split(X, y):
        model.fit(X[tr_i], y[tr_i])
        scores.append(f1_score(y[vl_i], model.predict(X[vl_i]), average='weighted'))
    return np.mean(scores)

def objective_lgbm(trial):
    params = dict(
        n_estimators      = trial.suggest_int('n_estimators', 300, 1200),
        max_depth         = trial.suggest_int('max_depth', 3, 10),
        num_leaves        = trial.suggest_int('num_leaves', 20, 150),
        learning_rate     = trial.suggest_float('learning_rate', 0.005, 0.2, log=True),
        subsample         = trial.suggest_float('subsample', 0.5, 1.0),
        colsample_bytree  = trial.suggest_float('colsample_bytree', 0.4, 1.0),
        reg_alpha         = trial.suggest_float('reg_alpha', 1e-8, 10.0, log=True),
        reg_lambda        = trial.suggest_float('reg_lambda', 1e-8, 10.0, log=True),
        min_child_samples = trial.suggest_int('min_child_samples', 5, 50),
        class_weight='balanced', random_state=42, n_jobs=-1, verbose=-1,
    )
    return cv_f1(LGBMClassifier(**params), X_tr, y_tr)

print(f'Tuning LightGBM ({N_TRIALS} trials, {CV_FOLDS}-fold CV)...')
study_lgbm = optuna.create_study(direction='maximize')
study_lgbm.optimize(objective_lgbm, n_trials=N_TRIALS, show_progress_bar=True)
print(f'\n Best LightGBM CV F1: {study_lgbm.best_value:.4f}')
print(f'   Best params: {study_lgbm.best_params}')
```

```python
# ── STEP 10: Train Models ─────────────────────────────────────────────────────

lgbm_params = dict(**study_lgbm.best_params, class_weight='balanced',
                   random_state=42, n_jobs=-1, verbose=-1)

print('Training LightGBM (tuned)...')
lgbm_m = LGBMClassifier(**lgbm_params)
lgbm_m.fit(X_tr, y_tr, eval_set=[(X_te,y_te)],
           callbacks=[lgb.early_stopping(40,verbose=False), lgb.log_evaluation(-1)])

print('Training LightGBM (deep)...')
lgbm_m2 = LGBMClassifier(n_estimators=800, max_depth=8, num_leaves=127,
                          learning_rate=0.03, subsample=0.7, colsample_bytree=0.7,
                          reg_alpha=0.1, reg_lambda=1.0, min_child_samples=20,
                          class_weight='balanced', random_state=99, n_jobs=-1, verbose=-1)
lgbm_m2.fit(X_tr, y_tr, eval_set=[(X_te,y_te)],
            callbacks=[lgb.early_stopping(40,verbose=False), lgb.log_evaluation(-1)])

print('Training Gradient Boosting...')
gb_m = GradientBoostingClassifier(n_estimators=300, max_depth=5, learning_rate=0.05,
                                   subsample=0.8, min_samples_leaf=10,
                                   max_features='sqrt', random_state=42)
gb_m.fit(X_tr, y_tr)

MODELS = {'LightGBM (tuned)':lgbm_m, 'LightGBM (deep)':lgbm_m2, 'GradientBoosting':gb_m}
print('\n All models trained')
```

```python
# ── STEP 11: Evaluate Models ──────────────────────────────────────────────────

print(f'{"Model":<22}  {"Train F1":>9}  {"Test F1":>9}')
print('-'*44)
for name, m in MODELS.items():
    print(f'{name:<22}  {f1_score(y_tr,m.predict(X_tr),average="weighted"):>9.4f}  '
          f'{f1_score(y_te,m.predict(X_te),average="weighted"):>9.4f}')

proba_te = np.mean([m.predict_proba(X_te) for m in MODELS.values()], axis=0)
ens_pred = proba_te.argmax(axis=1)
ens_f1   = f1_score(y_te, ens_pred, average='weighted')
print(f'{"Ensemble":<22}  {"—":>9}  {ens_f1:>9.4f}')
print(f'\nDetailed report (Ensemble):')
print(classification_report(y_te, ens_pred, target_names=le.classes_))

# Validation sanity check
val_preds = le.inverse_transform(
    np.mean([m.predict_proba(X_val) for m in MODELS.values()], axis=0).argmax(axis=1))
print('Validation distribution (sanity check):')
print(pd.Series(val_preds).value_counts())

cm = confusion_matrix(y_te, ens_pred)
ConfusionMatrixDisplay(cm, display_labels=le.classes_).plot(cmap='Blues')
plt.title(f'Ensemble Confusion Matrix — Test F1: {ens_f1:.4f}')
plt.grid(False); plt.tight_layout(); plt.show()

BEST_F1 = ens_f1
FINAL_MODELS = MODELS
X_val_final  = X_val
```

```python
# ── STEP 12: SHAP Feature Importance ─────────────────────────────────────────

print('Computing SHAP values...')
sample_idx  = np.random.choice(len(X_tr), size=min(2000,len(X_tr)), replace=False)
explainer   = shap.TreeExplainer(lgbm_m)
shap_values = explainer.shap_values(X_tr[sample_idx])

if isinstance(shap_values, list):
    mean_abs_shap = np.mean([np.abs(sv).mean(axis=0) for sv in shap_values], axis=0)
elif isinstance(shap_values, np.ndarray) and shap_values.ndim == 3:
    mean_abs_shap = np.abs(shap_values).mean(axis=(0,2))
else:
    mean_abs_shap = np.abs(shap_values).mean(axis=0)

mean_abs_shap = np.array(mean_abs_shap).ravel()
shap_df = (pd.DataFrame({'feature':FEATURE_COLS,'shap_importance':mean_abs_shap})
           .sort_values('shap_importance',ascending=False).reset_index(drop=True))

top40 = shap_df.head(40)
plt.figure(figsize=(12,10))
plt.barh(top40['feature'][::-1], top40['shap_importance'][::-1], color='steelblue')
plt.xlabel('Mean |SHAP Value|')
plt.title('Top 40 Features by SHAP Importance (LightGBM)')
plt.tight_layout(); plt.show()

print('\nTop 15 features:')
print(shap_df.head(15).to_string(index=False))

threshold = shap_df['shap_importance'].quantile(0.20)
KEEP_COLS = shap_df[shap_df['shap_importance']>threshold]['feature'].tolist()
print(f'\nKept {len(KEEP_COLS)}/{len(FEATURE_COLS)} features')
```

```python
# ── STEP 13: Retrain on SHAP-Selected Features ────────────────────────────────

keep_idx = [FEATURE_COLS.index(c) for c in KEEP_COLS]
X_tr2,X_te2,X_val2 = X_tr[:,keep_idx], X_te[:,keep_idx], X_val[:,keep_idx]
print(f'Retraining on {X_tr2.shape[1]} selected features...')

lgbm_s  = LGBMClassifier(**lgbm_params)
lgbm_s.fit(X_tr2,y_tr,eval_set=[(X_te2,y_te)],
           callbacks=[lgb.early_stopping(40,verbose=False),lgb.log_evaluation(-1)])
lgbm_s2 = LGBMClassifier(n_estimators=800,max_depth=8,num_leaves=127,learning_rate=0.03,
                          subsample=0.7,colsample_bytree=0.7,reg_alpha=0.1,reg_lambda=1.0,
                          min_child_samples=20,class_weight='balanced',
                          random_state=99,n_jobs=-1,verbose=-1)
lgbm_s2.fit(X_tr2,y_tr,eval_set=[(X_te2,y_te)],
            callbacks=[lgb.early_stopping(40,verbose=False),lgb.log_evaluation(-1)])
gb_s = GradientBoostingClassifier(n_estimators=300,max_depth=5,learning_rate=0.05,
                                   subsample=0.8,min_samples_leaf=10,
                                   max_features='sqrt',random_state=42)
gb_s.fit(X_tr2,y_tr)

MODELS2 = {'LightGBM (tuned)':lgbm_s,'LightGBM (deep)':lgbm_s2,'GradientBoosting':gb_s}
proba_te2 = np.mean([m.predict_proba(X_te2) for m in MODELS2.values()], axis=0)
ens_f1_2  = f1_score(y_te, proba_te2.argmax(axis=1), average='weighted')

print(f'Ensemble F1 (all features):    {ens_f1:.4f}')
print(f'Ensemble F1 (SHAP-selected):   {ens_f1_2:.4f}')

if ens_f1_2 >= ens_f1:
    FINAL_MODELS,X_val_final,BEST_F1 = MODELS2,X_val2,ens_f1_2
    print('→ Using SHAP-selected features.')
else:
    FINAL_MODELS,X_val_final,BEST_F1 = MODELS,X_val,ens_f1
    print('→ Full feature set performed better; using that.')

val_dist = pd.Series(le.inverse_transform(
    np.mean([m.predict_proba(X_val_final) for m in FINAL_MODELS.values()],axis=0).argmax(axis=1)
)).value_counts()
print(f'\nValidation distribution after Step 13:\n{val_dist}')
```

```python
# ── STEP 14: Pseudo-Labeling on Sierra Leone ──────────────────────────────────
# Adds high-confidence Sierra Leone predictions back to training.
# Distribution guard: rejects pseudo-labeled models if Medium < 5%.

CONFIDENCE_THRESHOLD = 0.82
proba_val = np.mean([m.predict_proba(X_val_final) for m in FINAL_MODELS.values()], axis=0)
max_prob  = proba_val.max(axis=1)
pseudo_y  = proba_val.argmax(axis=1)
conf_mask = max_prob >= CONFIDENCE_THRESHOLD
n_pseudo  = conf_mask.sum()
print(f'Pseudo-labels: {n_pseudo} samples >= {CONFIDENCE_THRESHOLD} confidence')
print(pd.Series(le.inverse_transform(pseudo_y[conf_mask])).value_counts())

if n_pseudo > 0:
    X_base = X_tr2 if FINAL_MODELS is MODELS2 else X_tr
    X_aug  = np.vstack([X_base, X_val_final[conf_mask]])
    y_aug  = np.concatenate([y_tr, pseudo_y[conf_mask]])
    print(f'Augmented training: {len(X_aug):,} rows')

    lgbm_pl  = LGBMClassifier(**lgbm_params)
    lgbm_pl.fit(X_aug, y_aug)
    lgbm_pl2 = LGBMClassifier(n_estimators=800,max_depth=8,num_leaves=127,
                               learning_rate=0.03,subsample=0.7,colsample_bytree=0.7,
                               reg_alpha=0.1,reg_lambda=1.0,min_child_samples=20,
                               class_weight='balanced',random_state=99,n_jobs=-1,verbose=-1)
    lgbm_pl2.fit(X_aug, y_aug)
    gb_pl = GradientBoostingClassifier(n_estimators=300,max_depth=5,learning_rate=0.05,
                                        subsample=0.8,min_samples_leaf=10,
                                        max_features='sqrt',random_state=42)
    gb_pl.fit(X_aug, y_aug)
    MODELS_PL = {'LightGBM (tuned)':lgbm_pl,'LightGBM (deep)':lgbm_pl2,'GradientBoosting':gb_pl}

    X_te_eval = X_te2 if FINAL_MODELS is MODELS2 else X_te
    f1_pl = f1_score(y_te,
        np.mean([m.predict_proba(X_te_eval) for m in MODELS_PL.values()],axis=0).argmax(axis=1),
        average='weighted')
    val_pl = pd.Series(le.inverse_transform(
        np.mean([m.predict_proba(X_val_final) for m in MODELS_PL.values()],axis=0).argmax(axis=1)
    )).value_counts()
    print(f'\nVal distribution after pseudo-labeling:\n{val_pl}')
    medium_pct = val_pl.get('Medium',0)/sum(val_pl)
    print(f'Test F1: {BEST_F1:.4f} → {f1_pl:.4f}')
    if f1_pl >= BEST_F1 and medium_pct > 0.05:
        FINAL_MODELS,BEST_F1 = MODELS_PL,f1_pl
        print('→ Using pseudo-labeled models.')
    else:
        reason = 'distribution skewed' if medium_pct <= 0.05 else 'F1 did not improve'
        print(f'→ Pseudo-labeling rejected ({reason}). Keeping prior models.')
else:
    print('No pseudo-labels added.')
```

```python
# ── STEP 15: Final Predictions for Sierra Leone ───────────────────────────────

final_proba  = np.mean([m.predict_proba(X_val_final) for m in FINAL_MODELS.values()], axis=0)
final_labels = le.inverse_transform(final_proba.argmax(axis=1))

coords = pd.read_csv(CSV_SIERRA)
submission = pd.DataFrame({
    'Longitude': coords['Longitude'].values,
    'Latitude':  coords['Latitude'].values,
    'UHI_Class': final_labels
})

print('Prediction distribution (Sierra Leone):')
counts = submission['UHI_Class'].value_counts()
print(counts)
for cls in ['High', 'Low', 'Medium']:
    pct  = counts.get(cls, 0) / len(submission)
    flag = ' very low' if pct < 0.05 else ''
    print(f'  {cls}: {pct:.1%}{flag}')

# Save to Colab local storage
submission.to_csv(OUTPUT_CSV, index=False)
print(f'\n Saved → {OUTPUT_CSV}')
display(submission.head(10))
```

```python
# ── ROBUST FIX: Temperature Scaling + Class Weight Correction ─────────────────
#
# Three-stage approach:
# Stage 1: Retrain with class weights matching ground truth distribution
# Stage 2: Calibrate probabilities using Platt Scaling (sigmoid calibration)
# Stage 3: Apply threshold optimisation per class

from sklearn.calibration import CalibratedClassifierCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
import warnings
warnings.filterwarnings('ignore')

# ── STAGE 1: Retrain with ground-truth-informed class weights ─────────────────
# Ground truth revealed: High=36%, Low=21%, Medium=43%
# LabelEncoder order:    High=0,   Low=1,   Medium=2

print("=" * 60)
print("STAGE 1: Retraining with corrected class weights")
print("=" * 60)

# Inverse frequency weighting based on ground truth distribution
gt_dist = {0: 0.361,   # High
           1: 0.214,   # Low
           2: 0.425}   # Medium

class_weights_gt = {k: 1.0 / v for k, v in gt_dist.items()}
print(f"Class weights → High: {class_weights_gt[0]:.2f}  "
      f"Low: {class_weights_gt[1]:.2f}  "
      f"Medium: {class_weights_gt[2]:.2f}")

# Retrain LightGBM tuned with corrected weights
lgbm_gt = LGBMClassifier(
    **{k:v for k,v in lgbm_params.items() if k != 'class_weight'},
    class_weight=class_weights_gt,
    random_state=42, n_jobs=-1, verbose=-1
)
lgbm_gt.fit(
    X_tr, y_tr,
    eval_set=[(X_te, y_te)],
    callbacks=[lgb.early_stopping(40, verbose=False),
               lgb.log_evaluation(-1)]
)

# Retrain LightGBM deep with corrected weights
lgbm_gt2 = LGBMClassifier(
    n_estimators=800, max_depth=8, num_leaves=127,
    learning_rate=0.03, subsample=0.7, colsample_bytree=0.7,
    reg_alpha=0.1, reg_lambda=1.0, min_child_samples=20,
    class_weight=class_weights_gt,
    random_state=99, n_jobs=-1, verbose=-1
)
lgbm_gt2.fit(
    X_tr, y_tr,
    eval_set=[(X_te, y_te)],
    callbacks=[lgb.early_stopping(40, verbose=False),
               lgb.log_evaluation(-1)]
)

# GradientBoosting with corrected weights
gb_gt = GradientBoostingClassifier(
    n_estimators=300, max_depth=5, learning_rate=0.05,
    subsample=0.8, min_samples_leaf=10, max_features='sqrt',
    random_state=42
)
# GBC doesn't support class_weight dict directly — use sample weights
sample_weights = np.array([class_weights_gt[yi] for yi in y_tr])
gb_gt.fit(X_tr, y_tr, sample_weight=sample_weights)

MODELS_GT = {
    'LightGBM (tuned)':  lgbm_gt,
    'LightGBM (deep)':   lgbm_gt2,
    'GradientBoosting':  gb_gt,
}

# Evaluate on test set
proba_gt = np.mean([m.predict_proba(X_te) for m in MODELS_GT.values()], axis=0)
f1_gt    = f1_score(y_te, proba_gt.argmax(axis=1), average='weighted')
print(f'\nTest F1 with corrected class weights: {f1_gt:.4f}')
print(f'Previous test F1:                     {BEST_F1:.4f}')


# ── STAGE 2: Platt Scaling calibration on held-out test set ───────────────────
print()
print("=" * 60)
print("STAGE 2: Platt Scaling probability calibration")
print("=" * 60)

# Fit a logistic regression on the raw probabilities from test set
# This learns how to map uncalibrated probs → calibrated probs
raw_proba_te = np.mean([m.predict_proba(X_te) for m in MODELS_GT.values()], axis=0)

platt = LogisticRegression(
    multi_class='multinomial', max_iter=1000,
    C=1.0, random_state=42
)
platt.fit(raw_proba_te, y_te)

# Apply calibration to validation set
raw_proba_val   = np.mean([m.predict_proba(X_val_final) for m in MODELS_GT.values()], axis=0)
cal_proba_val   = platt.predict_proba(raw_proba_val)

print("Distribution BEFORE calibration:")
before = le.inverse_transform(raw_proba_val.argmax(axis=1))
print(pd.Series(before).value_counts())

print("\nDistribution AFTER Platt calibration:")
after = le.inverse_transform(cal_proba_val.argmax(axis=1))
print(pd.Series(after).value_counts())


# ── STAGE 3: Per-class threshold optimisation ─────────────────────────────────
print()
print("=" * 60)
print("STAGE 3: Per-class threshold optimisation")
print("=" * 60)

# Use calibrated test set probabilities to find optimal thresholds
cal_proba_te = platt.predict_proba(raw_proba_te)

best_thresholds = {}
best_f1_thresh  = 0
best_combo      = None

# Grid search over threshold combinations
# Class order: High=0, Low=1, Medium=2
# We prioritise Medium recall since it was most under-predicted
thresholds_high   = np.arange(0.25, 0.65, 0.05)
thresholds_medium = np.arange(0.20, 0.55, 0.05)

print("Searching for optimal thresholds...")
for t_high in thresholds_high:
    for t_med in thresholds_medium:
        # Apply thresholds: predict class if prob exceeds threshold
        # If multiple exceed, take highest; if none exceed, take highest prob
        preds = []
        for probs in cal_proba_te:
            if probs[2] >= t_med:        # Medium threshold
                preds.append(2)
            elif probs[0] >= t_high:     # High threshold
                preds.append(0)
            else:
                preds.append(probs.argmax())
        f1_t = f1_score(y_te, preds, average='weighted')
        if f1_t > best_f1_thresh:
            best_f1_thresh = f1_t
            best_combo     = (t_high, t_med)

print(f"Best thresholds → High: {best_combo[0]:.2f}, Medium: {best_combo[1]:.2f}")
print(f"Best threshold F1 on test set: {best_f1_thresh:.4f}")


# ── FINAL: Apply everything to Sierra Leone ───────────────────────────────────
print()
print("=" * 60)
print("FINAL: Generating improved Sierra Leone predictions")
print("=" * 60)

t_high, t_med = best_combo
final_labels_robust = []
for probs in cal_proba_val:
    if probs[2] >= t_med:
        final_labels_robust.append(2)
    elif probs[0] >= t_high:
        final_labels_robust.append(0)
    else:
        final_labels_robust.append(probs.argmax())

final_labels_robust = le.inverse_transform(np.array(final_labels_robust))

print("Final prediction distribution:")
robust_counts = pd.Series(final_labels_robust).value_counts()
print(robust_counts)
for cls in ['High', 'Low', 'Medium']:
    pct  = robust_counts.get(cls, 0) / len(final_labels_robust)
    print(f"  {cls}: {pct:.1%}")

# Save
coords = pd.read_csv(CSV_SIERRA)
submission_robust = pd.DataFrame({
    'Longitude': coords['Longitude'].values,
    'Latitude':  coords['Latitude'].values,
    'UHI_Class': final_labels_robust
})

OUTPUT_ROBUST = OUTPUT_CSV.replace('Predicted_Dataset', 'Predicted_Dataset_Robust')
submission_robust.to_csv(OUTPUT_ROBUST, index=False)
print(f'\n Saved → {OUTPUT_ROBUST}')

from google.colab import files
files.download(OUTPUT_ROBUST)
```
